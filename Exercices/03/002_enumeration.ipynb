```python
import itertools
v = [4, 9, 0, 34, 7]
k = 3
```

```python
# Générer tous les sous ensembles de 3 éléments parmi v (n choose k)
for subset in itertools.combinations(v, k):
    print(subset)
```

```python
# Autre version
subsets = list(itertools.combinations(v, k))
print(subsets)
```

```python
# Avec répétition
for subset in itertools.combinations_with_replacement(v, k):
    print(subset)
```

```python
# Générer tous les subsets d'un ensemble (2^n)
for k in range(len(v)+1):
    for subset in itertools.combinations(v, k):
        print(subset)
```

```python
# Générer toutes les permutations (n!)
for perm in itertools.permutations(v):
    print(perm)
```

```python
# Tous les subsets de k éléments parmi n mais où l'odre a de l'importance (n!/(n-k)!)
for perm in itertools.permutations(v, k):
    print(perm)
```
