```python
import itertools
```

```python
liste1 = ['Audi', 'Peugeot', 'Opel']
liste2 = ['rouge', 'bleu']

for element in itertools.product(liste1, liste2):
    print(element)
```

```python
listes = [[1,2,3], [3,4,5], [7,8], [0,1,3,4]]

for element in itertools.product(*listes):
    print(element)
```

```python
for element in itertools.product([0,1], repeat=10):
    print(element)
```

```python
# EXERCICE 1
# Génèrer l'ensemble des mots de passe de 5 caractères sachant que
# les 2 premiers caractères sont des lettres (minuscules ou majuscules)
# et les trois derniers caractères sont des chiffres.

lettres = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
chiffres = "0123456789"

for part1 in itertools.product(lettres, repeat=2):
    for part2 in itertools.product(chiffres, repeat=3):
        print("".join(part1) + "".join(part2))
```

```python
lettres = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
chiffres = "0123456789"

# Utiliser islice pour limiter le nombre de résultats
passwords = ("".join(part1 + part2)
             for part1 in itertools.product(lettres, repeat=2)
             for part2 in itertools.product(chiffres, repeat=3))

# Afficher les 10 premiers éléments
for password in itertools.islice(passwords, 10):
    print(password)
```

```python
# EXERCICE 2
# Génèrer l'ensemble des mots de passe de 5 caractères sachant que
# les 3 premiers caractères sont parmi 'ABCD' (sans répétition)
# et les deux derniers sont parmi [5,6,7,8] (sans répétition).
for part1 in itertools.combinations('ABCD', 3):
    for part2 in itertools.combinations('5678', 2):
        print("".join(part1) + "".join(part2))
```
